## Problem 1:  (Continuous Probability, 5 pts) 

This problem is a classic probability problem called `Buffon's Needle.`  The experiment
involves a large number of needles 1 inch long, and a large piece of paper with parallel lines
drawn 1 inch apart. The needles are dropped onto the paper randomly, and we count the number
of needles which cross one of the lines.  

The analytical probability  that a needle crosses a line is

$$2/\pi \ =\ 0.6366197723675814...$$

(see Wikipedia for a derivation). 

In this problem, we will simulate this experiment, and use it to calculate $\pi$. Such methods of
calculation involving random trials are called "Monte Carlo" algorithms, in honor of the famous
casino, where gamblers to this day still show off their ignorance of probability.
